```python
import pandas as pd
import seaborn as sns
import os
%matplotlib inline
```

```python
working_dir = os.path.abspath('../')
```

```python
df1 = pd.read_csv((working_dir + '/output/lvl1.csv'), index_col = 0)
```

```python
sns.set_style("dark")
sns.set_context("poster")

g = sns.catplot(boxprops = dict(alpha = .6), height = 10, aspect = 3,
                x = "tissue compartment", y = "proportion unassigned",
                hue = "reference taxonomy",
                hue_order = ["greengenes", "silva", "greengenes_metaxa2",
                             "silva_metaxa2"], kind = "box", data = df1,
                fliersize = 0)
h = sns.swarmplot(x = "tissue compartment", dodge = True,
                  y = "proportion unassigned",
                  hue = "reference taxonomy",
                  hue_order = ["greengenes", "silva",
                               "greengenes_metaxa2", "silva_metaxa2"],
                  size = 3, data = df1, ax = g.ax, palette = "dark")
h.legend_.remove()
```

```python
df5 = pd.read_csv((working_dir + '/output/lvl5.csv'), index_col = 0)
```

```python
i = sns.catplot(boxprops = dict(alpha = .6), height = 10, aspect = 3,
                x = "tissue compartment", y = "proportion mitochondria",
                hue = "reference taxonomy",
                hue_order = ["greengenes", "silva", "greengenes_metaxa2",
                             "silva_metaxa2"], kind = "box", data = df5,
                fliersize = 0)
j = sns.swarmplot(x = "tissue compartment", dodge = True,
                  y = "proportion mitochondria",
                  hue = "reference taxonomy",
                  hue_order = ["greengenes", "silva",
                               "greengenes_metaxa2", "silva_metaxa2"],
                  size = 3, data = df5, ax = i.ax, palette = "dark")
j.legend_.remove()
```
