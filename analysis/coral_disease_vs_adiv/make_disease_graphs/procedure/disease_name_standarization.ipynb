```python
import pandas as pd
import numpy as np
```

```python
reference = pd.read_csv("../input/column_name_mapping.csv")
reference.set_index("Disease Name", inplace=True)
reference.drop(columns = "Source", inplace = True)
ref_dict = reference.to_dict()
```

```python
reference.head(3)
```

### FRRP

```python
pre_FRRP = pd.read_csv("../input/FRRP_disease_prevalence_by_species.csv")
FRRP_renamed = pre_FRRP.rename(columns = ref_dict["GCMP Name"])
FRRP_renamed.head(10)
```

### HICORDIS

```python
pre_HICORDIS = pd.read_csv("../input/HICORDIS_disease_prevalence_by_species.csv")
HICORDIS_renamed = pre_HICORDIS.rename(columns = ref_dict["GCMP Name"])
HICORDIS_renamed.head(10)
```

### Lamb et al.

```python
pre_Lamb = pd.read_csv("../input/Australia_Disease_Data.csv")
Lamb_renamed = pre_Lamb.rename(columns = ref_dict["GCMP Name"])
Lamb_renamed.head(10)
```

## Output to csv files

```python
FRRP_renamed.to_csv("../input/datasets_renamed/FRRP_renamed.csv", index = False)
HICORDIS_renamed.to_csv("../input/datasets_renamed/HICORDIS_renamed.csv", index = False)
Lamb_renamed.to_csv("../input/datasets_renamed/Lamb_et_al_renamed.csv", index = False)
```
