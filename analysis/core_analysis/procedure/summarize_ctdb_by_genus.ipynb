```python
ctdb_data = ctdb_data.loc[ctdb_data.loc[:,"trait_name"] == 'Growth rate']
ctdb_data
```

```python
ctdb_data["conversion_factor_to_mm_per_year"] = 1.0

growth_rate_units = set(ctdb_data.loc[ctdb_data.loc[:,"trait_name"] == 'Growth rate',"standard_unit"])
print(growth_rate_units)


# some conversions are a little confusing but
#growth in mm per x/12  years (i.e. per x months) * 12/x = growth per 1 year
conversion_dict = {'mm yr^-1':1.0,'mm month^-1':12.0,'mm per 4 months^-1':12.0/4.0,
                   'mm per 3 months^-1':12/3.0,'mm per 7 months^-1':12.0/7.0,
                   'mm per 6 months^-1':12.0/6.0,'mm per 8 months^-1':12.0/8.0,'mm d^-1':365.0}

for unit in growth_rate_units:
    ctdb_data.loc[ctdb_data.loc[:,"standard_unit"] == unit,"conversion_factor_to_mm_per_year"] = conversion_dict[unit]
```

```python
 
from string import whitespace
genus_and_species = ctdb_data["specie_name"].str.split(n=1,expand=True)
genus_and_species.columns = ["host_genus","species_suffix"]
ctdb_data = ctdb_data.merge(genus_and_species,left_index=True,right_index=True) 
ctdb_data

```

```python
ctdb_data["converted_value"] = ctdb_data.loc[:,"value"].astype(float) * ctdb_data.loc[:,"conversion_factor_to_mm_per_year"].astype(float)
```

```python
ctdb_data.loc[:,["host_genus","species_suffix","converted_value","conversion_factor_to_mm_per_year","value","replicates"]]
```

```python
#estimate minimal replicate numbers
ctdb_data["replicates"] = ctdb_data["replicates"].fillna(1.0)

per_genus_growth_rates = ctdb_data.groupby("host_genus")["converted_value"].mean()
per_genus_replicates = ctdb_data.groupby("host_genus")["replicates"].sum()
```

```python
df = per_genus_growth_rates.to_frame().merge(per_genus_replicates,left_index=True,right_index=True)
```

```python
df.columns = ["growth_rate_mm_per_year","growth_rate_replicates"]
df = df.sort_values('growth_rate_mm_per_year')
df = df[df["growth_rate_replicates"] >= 5.0]
df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

coral_data = merged_table
#x_column = 'dominance_skeleton'
#x_column = "growth_rate_mm_per_year"
#y_column = 'perc_dis'
#x_column = 'tissue_D_0__Bacteria___D_1__Proteobacteria___D_2__Gammaproteobacteria___D_3__Oceanospirillales___D_4__Endozoicomonadaceae___D_5__Endozoicomonas'
#x_column = 'tissue_D_0__Archaea___D_1__Thaumarchaeota___D_2__Nitrososphaeria___D_3__Nitrosopumilales___D_4__Nitrosopumilaceae___D_5__Candidatus_Nitrosopumilus'
x_column = 'mucus_D_0__Bacteria___D_1__Proteobacteria___D_2__Gammaproteobacteria___D_3__Alteromonadales___D_4__Pseudoalteromonadaceae___D_5__Pseudoalteromonas'
x_column = 'tissue_D_0__Bacteria___D_1__Proteobacteria___D_2__Alphaproteobacteria___D_3__Sphingomonadales___D_4__Sphingomonadaceae___D_5__Sphingobium'
y_column = 'growth_rate_mm_per_year'


x = list(coral_data[x_column])
y = list(coral_data[y_column])
labels = list(coral_data.index)
print(coral_data.columns)
replicates = list(coral_data["growth_rate_replicates"])

#filter x for nan values
x_filtered = []
y_filtered = []
labels_filtered = []
for i,x_val in enumerate(x):
    y_val = y[i]
    label = labels[i]
    if not pd.isnull(x_val) and not pd.isnull(y_val):
        x_filtered.append(x_val)
        y_filtered.append(y_val)
        labels_filtered.append(label)
    

df = pd.DataFrame({x_column:x_filtered,y_column:y_filtered},index=labels_filtered)
df = df.sort_values(x_column)

x_data = df[x_column]
y_data = df[y_column]

# Fit a regression line to the data
formula = f"{y_column} ~ {x_column}"
model = smf.ols(formula,data=df)
model_fit = model.fit()
prediction = model_fit.get_prediction()
model_prediction_df = prediction.summary_frame()

#Set up subplots
fig,ax = plt.subplots()

#We want scatterplot points and the thick line
#to plot on top of our shaded 95% confidence interval,
#so we plot the confidence interval first

#Plot a shaded area for the 95% prediction interval
#95% of future observed y values for a given x value should fall within this range
#Shade in the outermost confidence interval first, as it will be on bottom
lower_obs_ci = model_prediction_df["obs_ci_lower"]
upper_obs_ci = model_prediction_df["obs_ci_upper"]
ax.fill_between(x_data,lower_obs_ci,upper_obs_ci,alpha=0.1,color='grey')

#Add dotted lines to the edges of the 95% prediction interval
ax.plot(x_data,lower_obs_ci,alpha=0.75,marker=None,linestyle='dotted',color='grey',\
  label="95% Confidence Interval of Observations")
ax.plot(x_data,upper_obs_ci,alpha=0.75,marker=None,linestyle='dotted', color='grey')

#Plot a shaded area for the 95% confidence interval of the regression
#the true regression line (if the model is correct) should fall within this area 95% of the time
lower_ci = model_prediction_df["mean_ci_lower"]
upper_ci = model_prediction_df["mean_ci_upper"]
ax.fill_between(x_data,lower_ci,upper_ci,alpha=0.2,color='crimson')

#Add dashed lines to the edges of the 95% confidence interval
ax.plot(x_data,lower_ci,alpha=0.5,marker=None,linestyle='dashed',color='crimson',\
    label="95% Confidence Interval of Mean")
ax.plot(x_data,upper_ci,alpha=0.5,marker=None,linestyle='dashed',color='crimson')

#Get predicted values from the model
y_predicted = model_fit.fittedvalues
#Plot the model fit line using a thick black line
ax.plot(x_data,y_predicted,label=f"Regression line ({formula})",color="black",linewidth=2)


ax.scatter(x_data,y_data,color = "cyan",edgecolor="black",s= 100,alpha=0.5)

#Extract R2 from model and add to scatterplot
R2 = round(model_fit.rsquared,2)
ax.annotate(xy=(max(x_data)*0.5,max(upper_ci)*0.95),text=fr'$R^2 = {R2}$',size="xx-large")

#Extract p from the model and add to scatterplot
p = round(model_fit.f_pvalue,2)
ax.annotate(xy=(max(x_data)*0.5,max(upper_ci)*0.85),text=fr'$p = {p}$',size="xx-large")

plt.xlabel(f"{x_column}")
plt.ylabel(f"{y_column}")
```
